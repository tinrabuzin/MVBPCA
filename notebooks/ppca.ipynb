When the model is done fitting, we can plot the Hinton diagram of the weight matrix.

```python
def hinton(matrix, max_weight=None, ax=None):
    """Draw Hinton diagram for visualizing a weight matrix."""
    ax = ax if ax is not None else plt.gca()

    if not max_weight:
        max_weight = 2 ** np.ceil(np.log(np.abs(matrix).max()) / np.log(2))

    ax.patch.set_facecolor('gray')
    ax.set_aspect('equal', 'box')
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())

    for (x, y), w in np.ndenumerate(matrix):
        color = 'white' if w > 0 else 'black'
        size = np.sqrt(np.abs(w) / max_weight)
        rect = plt.Rectangle([x - size / 2, y - size / 2], size, size,
                             facecolor=color, edgecolor=color)
        ax.add_patch(rect)

    ax.autoscale_view()
    ax.invert_yaxis()
```

```python
hinton(vb_pca.q_w.mean[0].T)
C = np.identity(10) / vb_pca.q_tau.e_tau + vb_pca.q_w.mean[0] @ vb_pca.q_w.mean[0].T
```

The covariance matrix for the posterior distribution of the observed variables is close to the true value.

```python
print(np.diag(C))
```

There are only 4 column vectors left which are non-zero. The model automatically found the correct dimension of the latent space.

## 2D Example

First, genereate a bivariate distribution, plot it's p.d.f, and sample some points form it.

```python

X = np.arange(-5, 5, .1)
Y = np.arange(-2, 5, .1)
x, y = np.meshgrid(X, Y)
p = mlab.bivariate_normal(x, y, 1.5, 0.8, 1, 2, 0.6)
mean_pre = np.array([1,2])
cov_pre = np.array([[1.5**2, 0.6],[0.6, 0.8**2]])
sample = np.random.multivariate_normal(mean_pre, cov_pre, 10000).T

fig = plt.figure(figsize=(8, 8))
cs = plt.contour(x,y,p)
plt.scatter(sample[0], sample[1],alpha=0.7, s=2) # color=c, alpha=0.7, s=2)
plt.xlabel('Variable x')
plt.ylabel('Variable y')
plt.title('True Distribution and Sampled Data Points')
```

Now I run the algorithm which is provided the sampled data points and plot the results.

```python
vb_pca = mbpca.VBPCA(sample)
vb_pca.fit(max_iter=10000, min_iter=10, eps=1e-3, verbose=False, init=True)
```

```python
mean = vb_pca.q_mu.mean.T.reshape(2,1)
cov = vb_pca.q_w.mean[0] @ vb_pca.q_w.mean[0].T + np.identity(2) / vb_pca.q_tau.e_tau
p = mlab.bivariate_normal(x, y, np.sqrt(cov[0,0]), np.sqrt(cov[1,1]), mean[0], mean[1], cov[0,1])


fig = plt.figure(figsize=(8, 8))
cs = plt.contour(x,y,p)
plt.scatter(sample[0], sample[1],alpha=0.7, s=2) # color=c, alpha=0.7, s=2)
plt.xlabel('Variable x')
plt.ylabel('Variable y')
plt.title('Infered Distribution and Sampled Data Points')
```

We can check the posterior distribution which algorithm calculates, and the true distribution parameters.

```python
print('True mean:\n', mean_pre, '\n Calculated mean: \n', mean.reshape(2,))
```

```python
print('True covariance:\n', cov_pre, '\n Calculated covariance: \n', cov)
```

We can also check if the variational lower bound increases at every iteration.

```python
fig = plt.figure(figsize=(8, 8))
plt.plot(vb_pca.l_q[1:])
plt.xlabel('Iteration Number')
plt.ylabel('l(q)')
plt.title('Lower Bound During Training')
```
